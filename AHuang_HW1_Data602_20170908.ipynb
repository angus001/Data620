```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
G = nx.DiGraph()
G.add_node('John')
G.add_edge('John','Sylvie')
G.add_nodes_from(['Sylvie', 'Oliver'])
G.add_edges_from([('John','Oliver'),('Sylvie','Oliver')])
```

```python
G.add_nodes_from(['Andre','Heather','Ike','Jane','Ed','Diane'])
G.add_edges_from([('Ed','Diane'),('John','Diane'),('Diane','Andre'),('Ike','Heather')])
```

```python
nx.draw_circular(G, with_labels =True, node_size =2500)
plt.show()
```
